```python
import pandas as pd
from io import StringIO

# Sample data (you should replace this with your full data set)
data = """
Year,Country Name,Country Code,GDP per capita,Infant Mortality
1960,Bangladesh,BGD,84.83,174.8
1960,Cuba,CUB,NaN,42.8
1960,Nepal,NPL,49.99,217.9
1961,Bangladesh,BGD,92.86,170.7
1961,Cuba,CUB,NaN,41.3
1961,Nepal,NPL,51.32,215.2
1962,Bangladesh,BGD,95.05,166.9
1962,Cuba,CUB,NaN,39.8
1962,Nepal,NPL,54.31,211.9
"""

# Use StringIO to simulate reading from a file
df = pd.read_csv(StringIO(data))

# Ensure 'Country Name' is not part of the index
df.reset_index(drop=True, inplace=True)

# Convert data types
df['GDP per capita'] = pd.to_numeric(df['GDP per capita'], errors='coerce')
df['Infant Mortality'] = pd.to_numeric(df['Infant Mortality'], errors='coerce')

# Group by country and apply interpolation
df_grouped = df.groupby('Country Name').apply(lambda group: group.interpolate(method='linear'), include_groups=False)

# Reset index after groupby if necessary
df_grouped.reset_index(drop=True, inplace=True)

# Check results
df_grouped

```
